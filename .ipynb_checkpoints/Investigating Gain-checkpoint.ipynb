```python
import time
timing = []
def tic():
    timing.append(time.clock())
def toc():
    print 'elapsed: {:2f} seconds'.format(time.clock() - timing.pop())
    

def stats(data, others=dict()):
    if not isinstance(data, pd.DataFrame):
        data = data.to_frame()
    stats = pd.concat([data.nunique(), data.dtypes, data.isnull().sum()], axis=1)
    stats.columns = ['Unique', 'Dtypes', 'NaN Count']
    for k,v in others.items():
        stats[k] = v(data)
    return stats

def page(data, wrap_cols=14):
    if not isinstance(data, pd.DataFrame):
        data = data.to_frame().T
    for i in xrange(wrap_cols, data.shape[1]+wrap_cols, wrap_cols):
        print 'Columns', i-wrap_cols, '-', min(i-1,data.shape[1]-1)
        display(data.iloc[:, i-wrap_cols:i])
        

# %load ../dist_skew_plots.py
import seaborn as sns
from scipy.stats import skewtest, skew
from matplotlib import pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 20
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['font.size'] = 14

def setup_subplots (num_plots, plot_width = 20, sub_width=4, sub_height=4, hspace=.4, wspace=.3, sharey=False):
    n = num_plots
    cols = int(plot_width / sub_width)
    rows = int((n + cols - 1) / cols)

    fig, axes = plt.subplots(rows, cols, figsize=(plot_width,rows * sub_height), sharey=sharey)
    plt.subplots_adjust(hspace=hspace, wspace=wspace)
    for i in xrange(n - rows*cols, 0):
        plt.delaxes(axes.flat[i])
    return fig, axes

def dist_skew_plots(data, subplots=None, palette='gist_rainbow', show_palette=True, **distplot_kws):
    n = len(data.columns)
    colors = [(tuple(x**.3 for x in rgb), tuple(x**3 for x in rgb)) for rgb in sns.color_palette(palette, n_colors=n)]

    if show_palette:
        sns.palplot(sns.color_palette(palette, n_colors=n))

    if subplots is None:
        fig, axes = setup_subplots(n)
    else:
        fig, axes = subplots

    for ((rgb, dark_rgb), column, ax) in zip(colors, data.columns, axes.flat):
        if data[column].isnull().any():
            ax.text(.5, .5,
                ('{} had {} NaN\n'
                'values (of {}).\n'
                'Skipping.').format(
                    str(column).title(),
                    data[column].isnull().sum(),
                    len(data[column])),
                horizontalAlignment = 'center',
                verticalAlignment = 'center',
                color='r',
                fontsize=16,
                bbox = dict(facecolor='w', alpha=0.75),
                transform = ax.transAxes
                )
            ax.grid()
            ax.set_xticks([])
            ax.set_yticks([])
            continue

        options = {'kde': True,
                   'kde_kws': {'shade': True, 'color':dark_rgb},
                   'hist_kws': {'edgecolor': 'k'},}
        options.update(distplot_kws)
        sns.distplot(data[column], color=rgb, axlabel=False, ax=ax, **options)
        ax.set_title(str(column).title())
        ax.grid(axis='x')
        ax.set_yticks([])

        if data[column].nunique() == 2:
            continue
        else:
            stat, pvalue = skewtest(data[column])
            direction = 'left' if stat > 0 else 'right'
            x = 0.75 if stat > 0 else 0.25
            if pvalue < 0.05:
                ax.text(x, 0.75,
                    'Skewed\n{}: {:.2f}'.format(direction, skew(data[column])),
                    horizontalAlignment='center',
                    bbox=dict(facecolor='w', alpha=0.75),
                    transform=ax.transAxes)

    plt.show()
    

from scipy.stats import norm
import warnings

def histogram_flatten (data, cf_new_inverse=norm.ppf, max_bins=np.Inf, min_bins=100, verbose=False, save=False):
    bins = min(max(min(data.shape[0],min_bins),data.shape[0]/2), max_bins)
    if verbose:
        print 'Bins used:', bins

    hist, bins = np.histogram(data, bins=bins)
    if len(hist) < 3:
        return None
    cf = np.cumsum(hist)
    cf = cf / 1.0 / cf[-1]
    cf_new = cf_new_inverse(cf)    

    if cf_new[-1] == np.Inf:
        cf_new[-1] = cf_new[-2] + cf_new[-2] - cf_new[-3]

    if save:
        return np.interp(data, bins[:-1], cf_new), [cf_new, bins[:-1]]
    else:
        return np.interp(data, bins[:-1], cf_new)

def histogram_flatten_inverse (data, hist=None, bins=None):
    return np.interp(data, hist, bins)

from scipy.stats import boxcox as boxcox

def is_skewed(data, pvalue_cutoff=0.05):
    if len(np.unique(data)) <= 2:
        return False
    return skewtest(data).pvalue < pvalue_cutoff

def fourth_root(x):
    if any (x < 0):
        warnings.warn('\n\tNegative value encountered for fourth_root.'
                      '\n\tShifting by (minimum value) * 1.1 = ({}) * 1.1'.format(abs(x.min())))
        x += abs(x.min()) * 1.1
    return x**0.25

def histogram_full(x, **kwargs):
    return histogram_flatten(x,min_bins=np.Inf, **kwargs)

def inv_sqrt_p1(x):
    if any (x < 0):
        warnings.warn('\n\tNegative value encountered for inv_sqrt_p1.'
                      '\n\tShifting by (minimum value) * 1.1 = ({}) * 1.1'.format(abs(x.min())))
        x += abs(x.min()) * 1.1
    return 1.0/(1+x**0.5)

def inv_p1(x):
    return 1.0/(1+x)

def inv_1_5_p1(x):
    if any (x < 0):
        warnings.warn('\n\tNegative value encountered for inv_1_5_p1.'
                      '\n\tShifting by (minimum value) * 1.1 = ({}) * 1.1'.format(abs(x.min())))
        x += abs(x.min()) * 1.1
    return 1.0/(1+x**1.5)

def inv_sq_p1(x):
    return 1.0/(1+x**2)

def box_cox(x, save=False):
    if np.min(x) <= 0:
        shift = abs(np.min(x)) + 1
        x += shift
        result = [e for e in boxcox(x)] + [shift]
    else:
        result = boxcox(x)
    if save:
        return result
    else:
        return result[0]

def inv_box_cox(x, lam=None, shifted_by=1):
    if lam == 0:
        result = np.exp(x)
    else:
        result = (lam*x + 1)**(1.0/lam)
    if shifted:
        return result - shifted_by
        

def pick_unskew(data, transforms=[], verbose=False, show_skew=False, return_skews=False, inverting=False):
    transforms = np.unique([np.sqrt,
                            np.cbrt,
                            fourth_root,
                            np.log1p,
                            histogram_flatten,
                            histogram_full,
                            inv_sqrt_p1,
                            inv_p1,
                            inv_1_5_p1,
                            inv_sq_p1,
                            box_cox
                           ] + transforms).tolist()
    
    flip = skew(data) < 0
    if flip:
        data = max(data) - data
    
    skews = [skew(f(data)) for f in transforms]
    best = np.argmin(np.abs(skews))
    
    if verbose:
        print data.name
        if show_skew:
            print 'Skews:', {func:skew for func,skew in zip(transforms,skews)}
        print 'Transform Picked:', transforms[best]
        print 'Flipped:', flip
        print
    
    if inverting and transforms[best] in (histogram_flatten, histogram_full):
        _, [hist, bins] = transforms[best](data, save=True)
        if not return_skews:
            return transforms[best], flip, hist, bins
        else:
            return transforms[best], flip, skews, hist, bins
    
    if not return_skews:
        return transforms[best], flip
    else:
        return transforms[best], flip, skews
    

def unskew(data, transforms=[], **kwargs):
    transform, flip = pick_unskew(data, transforms, **kwargs)
    
    if flip:
        return pd.Series(transform(max(data) - data), index=data.index, name=data.name)
    else:
        return pd.Series(transform(data), index=data.index, name=data.name)

    
def reskew(data, inverse, flip=False, **kwargs):
    if inverse == np.sqrt:
        inverse = lambda x: x**2
    elif inverse == np.cbrt:
        inverse = lambda x: x**3
    elif inverse == fourth_root:
        inverse = lambda x: x**4
    elif inverse == np.log1p:
        inverse = np.expm1
    elif inverse in (histogram_flatten, histogram_full):
        inverse = histogram_flatten_inverse
    elif inverse == inv_sqrt_p1:
        inverse = lambda x: 1.0/(x)**2 - 1
    elif inverse == inv_p1:
        inverse = lambda x: 1.0/(x) - 1
    elif inverse == inv_1_5_p1:
        inverse = lambda x: 1.0/(x)**.67 - 1
    elif inverse == inv_sq_p1:
        inverse = lambda x: 1.0/(x)**0.5 - 1
    elif inverse == box_cox:
        inverse = inv_box_cox
                            
    result = inverse(data, **kwargs)
    if not flip:
        return pd.Series(result, index=data.index, name=data.name)
    else:
        return pd.Series(max(result) - result, index=data.index, name=data.name)
```

## A note about histogram flattener

This is a technique from image processing.
It is used in that context to equalize the intensity of pixels in an image (effectively increasing the contrast).
It is called histogram flattening because it transforms the distribution of intensities to be uniform (hence "flat").

Here, it is not used to "flatten" a feature's distribution to be uniform but rather to be a distribution of choice,
for instance a normal distribution.

The way to do this is essentially to map each group of samples with the same value $x$ to a new value $x'$ such that the
cumulative distribution of the transformed values is exactly the cumulative distribution of the desired distribution.

If we call each point in the original cumulative distribution function $(x,y)$ such that $x$ is the feature value and $y$ is the
cumulative distribution's value at $x$ (i.e. $cf(x) = y$ where $cf$ denotes the original cumulative distribution function), then our transform $T$ changes the sample values so that the new cumulative distribution $cf_{new}$ matches that of the desired target distribution $f$ and is defined as the $T(y) = T(cf(x)) = cf_{new}^{-1}(y)$.

The inverse of our transform exists because $cf_{new}^{-1}(y)$ is the inverse of the function $cf_{new}()$ and hence must be one-to-one.
In fact the inverse is the same as performing the transform from the desired distribution to the original distribution $T^{-1}(y) = T^{-1}(cf_{new}(x)) = cf^{-1}(y)$. As a practical note, since $cf$ is just the original cumulative distribution of the data obtained, it must be saved if its inverse is to be remembered after modifying the data with $T$.

```python
def get_gains(data, labels, target_value):
    N = data.shape[0]
    
    unique, inverse, count = np.unique(data, return_inverse=True, return_counts=True)
    freq = np.bincount(inverse[labels == target_value], minlength=len(unique))
    ratio = freq.sum() /1.0/ N
    gain = (freq - count * ratio) /1.0/ N
    
    return pd.Series(gain, index=unique)

def get_xtick_labels(index):
    if index.dtype == np.float:
        xtick_labels = ['{:.2f}'.format(i) for i in index]
    else:
        xtick_labels = index
        
    return xtick_labels

def is_valid_group (total, group_validation, feature):
    for i, g in enumerate(group_validation[feature]):
        if total < g:
            plt.text(.5, .5,
                ('Gain of {:.4f}\n'
                 'less than {:.4f}\n'
                 'of {}.\n'
                'Not Significant.').format(
                    total,
                    g,
                    feature.split('_')[i]
                    ),
                horizontalAlignment = 'center',
                verticalAlignment = 'center',
                color='w',
                fontsize=16,
                bbox = dict(facecolor='tab:gray', alpha=0.75),
                transform = plt.gca().transAxes
                )
            plt.grid()
            plt.xticks([])
            plt.yticks([])
            return False
        
    return True

def gain_plot (data, labels, target_value=0,
               scatter=False,
               setup=None,
               cmap='viridis',
               vmax=0.02,
               group_validation=None,
               gain_threshhold=0, #0.00002, # 1/50000, zeroed if gain is below this value
               gain_highlight=0.01, #0.001,
               return_gains=False,
               return_total=True,
               return_significant=True,
               show_significant=True,
               highlight_fontsize=24,
               xticks=8, xticks_fontsize=12,
               **setup_subplots_kws):

    D = data.shape[1]
    if return_gains:
        save_gains = {}
    if return_total:
        save_total = {}
    if return_significant:
        save_significant = []
    
    if setup is None:
        fig, axes = setup_subplots(D, sharey=True, **setup_subplots_kws)
    else:
        fig, axes = setup        
    
    for i, feature, ax in zip(range(D), data.columns, axes.flat):
        plt.sca(ax)
        
        plt.grid(b=True, axis='both')
        plt.yticks()
        
        gains = get_gains(data[feature], labels, target_value)
        gains = gains.apply(lambda x: np.nan if abs(x) < gain_threshhold else x)
        total = abs(gains).sum()
        
        if return_gains:
            save_gains[feature] = gains
        if return_total:
            save_total[feature] = total
        
        N = len(gains)
        x = range(N)

        if scatter:
            plt.scatter(x, gains, alpha=0.5, s=100000*abs(gains), c=abs(gains), cmap=cmap, vmin=0, vmax=vmax)
        else:
            plt.plot(x, gains)
        plt.fill_between(x, 0, gains, color='g', alpha=.25)

        if abs(gains.index).max() > 100 or gains.index.dtype == np.float:
            rotation = -45
        else:
            rotation = 0
        
        if N > xticks:
            xtick_labels = get_xtick_labels(gains[::N/xticks].index)
            plt.xticks(x[::N/xticks], xtick_labels, rotation=rotation, fontsize=xticks_fontsize)
        else:
            xtick_labels = get_xtick_labels(gains.index)
            plt.xticks(x, xtick_labels, rotation=rotation, fontsize=xticks_fontsize)

        if i % axes.shape[1] == 0:
            plt.ylabel('Gain\n(in Probability)')

        message = 'Gain: {:.4f}'.format(total)
        if data[feature].dtype == np.float:
            at = np.argmax(abs(gains))
            if np.isnan(at):
                message += '\nAll gains below' '\nthreshold.'
            else:
                max_gain = abs(gains[at])
                message += '\nMax Gain: {:.4f}\n'.format(max_gain) + '@ {:.2f}'.format(at)
        
        if total >= gain_highlight:
            if group_validation is not None \
            and not is_valid_group(total, group_validation, feature):
                continue
                
            if return_significant:
                save_significant.append(feature)
            
            plt.title(feature, color='g', fontsize=highlight_fontsize)
            ax.text(0.75, 0.75,
                message,
                horizontalAlignment='center',
                verticalAlignment='center',
                color='w',
                fontsize=16,
                bbox=dict(facecolor='g', alpha=0.75),
                transform=ax.transAxes)
        else:
            plt.title(feature)
            ax.text(0.75, 0.75,
                message,
                horizontalAlignment='center',
                verticalAlignment='center',
                bbox=dict(facecolor='w', alpha=0.75),
                transform=ax.transAxes)
            
    plt.show()
    
    result = []
    
    if return_gains:
        result.append(save_gains)
    if return_total:
        result.append(save_total)
    if return_significant:
        result.append(save_significant)
        if show_significant:
            print 'Significant Features ({} out of {}):\n'.format(len(save_significant), D), save_significant
    
    if result == []:
        pass
    elif len(result) == 1:
        return result[0]
    else:
        return result
```

```python
import xgboost as xgb
import pickle
import os.path
import operator

def print_divider(char='- ', n=None, lines=1, pre=True, post=True):
    if n is None:
        n = 100 / len(char)
    if pre: print
    for i in xrange(lines):
        print char*n
    if post: print

class XGB_GridSearcher:
    '''
    Does a grid search over boost_params_grid.
    Saves the results.
    Makes submission CSV file.
    Guards against overwriting files.
    Can load and save previous grid search results from a filepath.
    Written very specifically for binary classification with AUC evaluation metric.
    '''
    
    summary_cols_ = [
        'Underestimate Test', 'Overestimate Test',
         'test-auc-mean', 'test-auc-std',
         'train-auc-mean', 'train-auc-std',
         'Best Iteration'
    ]
    
    def __init__ (self, boost_params_grid=None, metadata=None, results=None, data=None, labels=None, filepath=None):
        if filepath is not None:
            self.load_from(filepath)
        else:
            self.boost_params_grid_ = boost_params_grid
            self.metadata_ = metadata
            if results is not None:
                self.results_ = results
            else:
                self.results_ = pd.DataFrame(columns = self.summary_cols_ + self.boost_params_grid_.keys())
            self.data_ = data
            self.labels_ = labels
    
    def load_from (self, filepath):
        with open(filepath, 'r') as f:
            other = pickle.load(f)
            
        self.__init__(other.boost_params_grid_, other.metadata_, other.results_, other.data_, other.labels_)

    @staticmethod
    def guard_existing (filepath):
        if os.path.isfile(filepath):
            raise Exception('''Error, don't override the existing file {}.'''.format(os.path.basename(filepath)))
    
    def pickle_me (self, filepath, save_data=True, verbose=True, prefix='gridsearch'):
        filepath = prefix + '/' + filepath
        self.guard_existing(filepath)
        
        if not save_data:
            tmp_data = self.data_
            tmp_labels = self.labels_
            self.data_ = None
            self.labels_ = None
        
        if verbose:
            print 'Pickling...'
            if save_data:
                print 'Including dataset as well...'
            tic()
        with open(filepath, 'w') as f:
            pickle.dump(self, f)
        if verbose: toc()
        
        if not save_data:
            self.data_ = tmp_data
            self.labels_ = tmp_labels

    def results_to_csv(self, filepath, prefix='gridsearch', index=False, **kwargs):
        filepath = prefix + '/' + filepath
        self.guard_existing(filepath)
        self.results_.to_csv(filepath, index=index, **kwargs)
    
    def search (self, data, labels, first=None, save_data=True, **cv_params):
        global timing
        
        self.cv_params_ = cv_params
        d_data = xgb.DMatrix(data, label=labels)
        
        tic()
        grid = self.boost_params_grid_.values()
        count = reduce(operator.mul, map(len, grid), 1)
        iterations = 0
        
        if first is not None and first < count:
            print 'Commencing search over the first {} of {} parameter combinations . . .\n'.format(first, count)
        else:
            print 'Commencing search over grid of {} parameter combinations . . .\n'.format(count)
        
        for i, param_vector in enumerate(itertools.product(*grid)):
            if first is not None and i == first:
                break
            
            now = time.clock() - timing[-1]
            print 'Combination {} of {}. (Time elapsed: {:.1f} seconds.)'.format(i+1, count, now)
            print 'Iterations Complete: {}.'.format(iterations)
            print 'Average rate: {:.1f} iterations/sec.'.format(iterations/1.0/now)
            boost_params = {k:v for k,v in zip(self.boost_params_grid_.keys(), param_vector)}
            result = pd.DataFrame({k:[v] for k,v in boost_params.items()}, index=['Params'])
            display(result)
            
            tic()
            history = xgb.cv(boost_params, d_data, **cv_params)
            toc()

            iterations += history.shape[0]
            result['Best Iteration'] = history.shape[0] - 1
            result = result.assign(**history.iloc[-1].to_dict())
            result['Underestimate Test'] = result['test-auc-mean'] - 2*result['test-auc-std']
            result['Overestimate Test'] = result['test-auc-mean'] + 2*result['test-auc-std']
            result = result[self.results_.columns]
            display(result[self.summary_cols_].rename(lambda x: 'Summary'))
            
            print_divider()
        
            self.results_ = self.results_.append(result, ignore_index=True)
        toc()
        
        if save_data:
            self.data_ = data.copy()
            self.labels_ = labels.copy()
        
        return self.results_
    
    def get_summary (self, index, transpose=True):
        if transpose:
            return self.results_.loc[index, self.summary_cols_].to_frame().T
        else:
            return self.results_.loc[index, self.summary_cols_].to_frame()
    
    def top_results (self, by='Underestimate Test', first=10, **kwargs):
        return self.results_.sort_values(by, ascending=False, **kwargs).head(first)
    
    def get_boost_params (self, index):
        return self.results_.loc[index, self.results_.columns.difference(self.summary_cols_)].to_dict()
    
    def get_best_iteration (self, index):
        return self.results_.loc[index, 'Best Iteration']
    
    def fit (self, index, data = None, labels = None, verbose=True):
        if data is None:
            data = self.data_
        if labels is None:
            labels = self.labels_
            
        if verbose:
            print 'Fitting...'
            display(self.results_.loc[index].to_frame().T)
            tic()
            
        result = xgb.train(
            self.get_boost_params(index),
            xgb.DMatrix(data, label=labels),
            num_boost_round = self.get_best_iteration(index)
        )
        
        if verbose:
            toc()
            
        return result
    
    def fit_predict (self, index, test, train = None, labels = None, verbose=True):
        if train is None:
            train = self.data_
        if labels is None:
            labels = self.labels_
            
        dtrain = xgb.DMatrix(train, label = labels)
        dtest = xgb.DMatrix(test)
        bst = self.fit(index, train, labels, verbose=verbose)
        if verbose:
            print 'Predicting...'
        return bst.predict(dtest)
    
    def make_submission (self, filepath, index, test, train = None, labels = None,
                         prefix='submissions',
                         index_label='id',
                         grid_name=None,
                         by='Underestimate Test',
                         verbose=True,
                         **kwargs):
        
        if train is None:
            train = self.data_
        if labels is None:
            labels = self.labels_
        
        filepath = prefix + '/' + filepath
        self.guard_existing(filepath)
        self.guard_existing(filepath + '.txt')
        
        preds = pd.DataFrame({'Y': self.fit_predict(index, test, train, labels, verbose=verbose)}, index = test.index)
        
        if verbose:
            print 'Making submission file: {}'.format(filepath)
        
        preds.to_csv(filepath, index_label=index_label, **kwargs)
        
        with open(filepath + '.txt', 'w') as f:
            f.write(str(self.get_summary(index, transpose=False)))
            f.write('\n\n')
            f.write('Metadata:\n' + str(self.metadata_))
        
        if grid_name is not None:
            print '(grid: {}) (id: {}) ({:.5f})'.format(grid_name, index, self.get_summary(index).loc[index, by])
```
